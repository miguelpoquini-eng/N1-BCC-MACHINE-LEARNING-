## 📦 Artefato 1: Coleta e Engenharia de Dados Sólida
Nesta etapa expandida, geramos interações realistas de **150 usuários** únicos cruzados com **12 categorias de produtos** de vestuário, forçando um percentual de dados nulos para simular cenários reais de produção imperfeitos.

```python
usuarios = [f"User_{i}" for i in range(1, 151)]
produtos = [
    "Camiseta Algodão", "Calça Jeans Slim", "Jaqueta de Couro", "Vestido Floral",
    "Moletom Canguru", "Tênis Casual", "Bermuda Sarja", "Blusa de Lã",
    "Saia Midi", "Camisa Linho", "Relógio Esportivo", "Bota de Cano Alto"
]

linhas = []
for u in usuarios:
    # Cada usuário interage com um subconjunto de itens (densidade controlada)
    prods_escolhidos = np.random.choice(produtos, size=np.random.randint(5, 10), replace=False)
    for p in prods_escolhidos:
        linhas.append({"usuario": u, "produto": p, "score_interesse": np.random.randint(1, 6)})

df_interacoes = pd.DataFrame(linhas)

# Inserindo 25 valores nulos propositais para testar nosso tratamento de dados
idx_nulos = np.random.choice(df_interacoes.index, 25, replace=False)
df_interacoes.loc[idx_nulos, "score_interesse"] = np.nan

print(f"📊 Dataset gerado com {len(df_interacoes)} registros.")
print(f"🔍 Quantidade de valores nulos iniciais: {df_interacoes['score_interesse'].isnull().sum()}")
```

```python
# Tratamento de nulos via imputação estatística pela Mediana
mediana = df_interacoes["score_interesse"].median()
df_interacoes["score_interesse"] = df_interacoes["score_interesse"].fillna(mediana)

# Normalização de Escala [0, 1] para o algoritmo de similaridade
scaler = MinMaxScaler()
df_interacoes["score_normalizado"] = scaler.fit_transform(df_interacoes[["score_interesse"]])

print(f"✅ Tratamento concluído. Valores nulos atuais: {df_interacoes['score_interesse'].isnull().sum()}")
display(df_interacoes.head())
```

## 🔀 Artefato 2: Divisão Robusta dos Dados e K-Fold
Para mitigar o vazamento de dados (*data leakage*), dividimos nossa população de usuários usando a estratégia **Holdout (80% treino / 20% teste)**. Adicionalmente, aplicamos **K-Fold com 5 splits** no conjunto de treino para mensurar a variabilidade da consistência interna.

```python
# Pivotagem de dados: Linhas=Usuários, Colunas=Produtos
matriz_completa = df_interacoes.pivot_table(
    index="usuario", columns="produto", values="score_normalizado"
).fillna(0)

usuarios_lista = matriz_completa.index.tolist()
usuarios_treino, usuarios_teste = train_test_split(
    usuarios_lista, test_size=0.20, random_state=42
)

matriz_treino = matriz_completa.loc[usuarios_treino]
matriz_teste  = matriz_completa.loc[usuarios_teste]

print(f"👥 Treinamento: {len(usuarios_treino)} usuários | Teste: {len(usuarios_teste)} usuários.")

# Execução do K-Fold (5 dobras) para validação interna da estabilidade
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores_cv = []

for fold, (tr_idx, val_idx) in enumerate(kf.split(matriz_treino), start=1):
    sim_fold = cosine_similarity(matriz_treino.iloc[tr_idx].T)
    scores_cv.append(sim_fold.mean())
    print(f" 🔹 Dobra {fold} - Similaridade Média Intra-Fold: {sim_fold.mean():.4f}")

print(f"\n📊 Estabilidade Geral Média do Modelo (CV): {np.mean(scores_cv):.4f}")
```

## ⚙️ Artefato 3: Treinamento do Modelo de Filtragem Colaborativa Item-Item
Utilizando a métrica matemática de **Similaridade Cosseno**, calculamos a correlação vetorial entre os produtos com base nas avaliações dos usuários mapeados no conjunto de treino. Produtos consumidos de forma concomitante adquirem maior peso cosseno.

```python
# Treinamento da similaridade cosseno
similaridade_itens = cosine_similarity(matriz_treino.T)

df_similaridade = pd.DataFrame(
    similaridade_itens,
    index=matriz_treino.columns,
    columns=matriz_treino.columns
)

print("✅ Matriz de Similaridade Item-Item computada com sucesso.")
print(f"Formato da Matriz de Pesos: {df_similaridade.shape}")
```

## 🚀 Artefato 4: Lógica do Motor de Recomendações
Implementação da função preditiva que analisa os itens com os quais o usuário já interagiu, multiplica-os pelas suas respectivas similaridades com os demais itens da loja e pontua os produtos não conhecidos de maior relevância.

```python
def recomendar_roupas(usuario, matriz_usuarios, df_sim, top_n=3):
    if usuario not in matriz_usuarios.index:
        return pd.Series(dtype=float)

    perfil_usuario = matriz_usuarios.loc[usuario]
    itens_vistos = perfil_usuario[perfil_usuario > 0].index
    pontuacao_recomendacoes = pd.Series(dtype=float)

    for item in itens_vistos:
        peso_usuario = perfil_usuario[item]
        itens_similares = df_sim[item] * peso_usuario
        pontuacao_recomendacoes = pontuacao_recomendacoes.add(itens_similares, fill_value=0)

    # Filtragem para não sugerir itens que o cliente já comprou/visualizou anteriormente
    pontuacao_recomendacoes = pontuacao_recomendacoes.drop(itens_vistos, errors="ignore")
    return pontuacao_recomendacoes.sort_values(ascending=False).head(top_n)

# Executando um teste piloto para o User_12
usuario_teste_piloto = "User_12"
sugestoes = recomendar_roupas(usuario_teste_piloto, matriz_completa, df_similaridade, top_n=3)

print(f"🛍️ Recomendações geradas para o {usuario_teste_piloto}:")
for i, (item, score) in enumerate(sugestoes.items(), 1):
    print(f"  {i}º Lugar: {item} (Score Preditivo: {score:.4f})")
```

## 📈 Artefatos 5 e 6: Métricas Estatísticas Completas
Binarizamos o problema assumindo que interações com score normalizado **>= 0.5** são itens realmente relevantes. Avaliamos o conjunto de testes (20% isolados) calculando Acurácia, Precisão, Recall e F1-Score para validar a capacidade preditiva industrial do motor.

```python
LIMIAR_RELEVANCIA = 0.5
TOP_K = 3
y_true, y_pred = [], []

for user in usuarios_teste:
    historico_real = matriz_completa.loc[user]
    predicoes_motor = recomendar_roupas(user, matriz_treino, df_similaridade, top_n=TOP_K)
    itens_sugeridos = set(predicoes_motor.index)

    for item in matriz_completa.columns:
        y_true.append(1 if historico_real[item] >= LIMIAR_RELEVANCIA else 0)
        y_pred.append(1 if item in itens_sugeridos else 0)

y_true, y_pred = np.array(y_true), np.array(y_pred)

# Geração das Métricas de Classificação no terminal
print("======================================================")
print("📊 RELATÓRIO DE MÉTRICAS DE MACHINE LEARNING (TESTE)")
print("======================================================")
print(classification_report(y_true, y_pred, target_names=["Não Relevante", "Relevante"], zero_division=0))

# Plotagem Gráfica Avançada (Heatmap Cosseno + Distribuição de Métricas)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(df_similaridade.iloc[:6, :6], annot=True, fmt=".2f", cmap="magma", ax=axes[0])
axes[0].set_title("Heatmap de Similaridade Cosseno (Amostra de 6 Itens)")

cm = confusion_matrix(y_true, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Não Relevante", "Relevante"]).plot(cmap="Blues", ax=axes[1])
axes[1].set_title("Matriz de Confusão do Sistema")
plt.tight_layout()
plt.show()
```

#### 🔹 Pontos Fortes da Arquitetura
* **Alta Explicabilidade:** O modelo é totalmente transparente, permitindo rastrear matematicamente o motivo pelo qual um produto foi sugerido (ex: *"Usuário recebeu a indicação da Jaqueta de Couro porque comprou a Bota de Cano Alto, que possui alta similaridade cosseno"*).
* **Escalabilidade Computacional:** Em e-commerces reais, o catálogo de produtos muda com muito menos frequência do que a entrada de novos clientes. Manter uma matriz Item-Item é computacionalmente mais barato do que recalcular uma matriz Usuário-Usuário.
* **Agregado de Metadados:** O sistema funciona perfeitamente registrando apenas o comportamento de consumo, sem a necessidade de criar tags manuais ou descrições complexas para cada peça de roupa.

#### 🔹 Limitações Identificadas e Soluções Futuras
1. **Problema do Cold Start (Início Frio):** O modelo atual é incapaz de gerar recomendações assertivas para usuários recém-cadastrados ou para produtos que acabaram de ser adicionados ao estoque, pois ambos não possuem histórico de interação. *Solução Futura:* Evolver a arquitetura para um Sistema de Recomendação Híbrido (*Hybrid Filtering*), injetando regras baseadas em conteúdo (*Content-Based*) para novos elementos.
2. **Esparsidade da Matriz:** À medida que a loja expande seu catálogo para milhares de itens, a matriz de interações tende a ficar cheia de valores zerados, prejudicando o cálculo de similaridade. *Solução Futura:* Implementar algoritmos de Decomposição de Matrizes, como SVD (*Singular Value Decomposition*) ou NMF, para compressão dimensional.
